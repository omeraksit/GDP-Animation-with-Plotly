# Plotly's Code

```python
url = "https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"
dataset = pd.read_csv(url)

years = ["1952", "1962", "1967", "1972", "1977", "1982", "1987", "1992", "1997", "2002",
         "2007"]

# make list of continents
continents = []
for continent in dataset["continent"]:
    if continent not in continents:
        continents.append(continent)
# make figure
fig_dict = {
    "data": [],
    "layout": {},
    "frames": []
}

# fill in most of layout
fig_dict["layout"]["xaxis"] = {"range": [30, 85], "title": "Life Expectancy"}
fig_dict["layout"]["yaxis"] = {"title": "GDP per Capita", "type": "log"}
fig_dict["layout"]["hovermode"] = "closest"
fig_dict["layout"]["updatemenus"] = [
    {
        "buttons": [
            {
                "args": [None, {"frame": {"duration": 500, "redraw": False},  #time between consecutive frames
                                "fromcurrent": True, "transition": {"duration": 4300,   #??
                                                                    "easing": "quadratic-in-out"}}],
                "label": "Play",
                "method": "animate"
            },
            {
                "args": [[None], {"frame": {"duration": 0, "redraw": False},
                                  "mode": "immediate",
                                  "transition": {"duration": 0}}],
                "label": "Pause",
                "method": "animate"
            }
        ],
        "direction": "left",    #the buttons will appear in left-to-right order 
        "pad": {"r": 10, "t": 87},   #Sets the padding around the buttons or dropdown menu. right, left, top, bottom
        "showactive": False,   #Highlights active dropdown item or active button if True. Default True.
        "type": "buttons",
        "x": 0.1,   #Type: number between or equal to -2 and 3,  Default: -0.05,  Sets the x position (in normalized coordinates) of the update menu.
        "xanchor": "right",  #one of ( "auto" | "left" | "center" | "right" )  Default "right".
        "y": 0,    #Type: number between or equal to -2 and 3,  Default: 1,   Sets the y position (in normalized coordinates) of the update menu.
        "yanchor": "top"   #one of ( "auto" | "top" | "middle" | "bottom" )  Default "top".
    }
]

sliders_dict = {
    "active": 0,   #Determines which button (by index starting from 0) is considered active. Default 0.
    "yanchor": "top",   #one of ( "auto" | "top" | "middle" | "bottom" )  Default "top".
    "xanchor": "left",   #one of ( "auto" | "left" | "center" | "right" )  Default "left".
    "currentvalue": {
        "font": {"size": 20},
        "prefix": "Year:",
        "visible": True,
        "xanchor": "right"   #one of ( "left" | "center" | "right" )  Default "left".
    },
    "transition": {"duration": 300, "easing": "cubic-in-out"},   #Duration of the slider transition. Default 150. / Default "cubic-in-out".
    "pad": {"b": 10, "t": 50},   #Default values: 't':20, 'r':0, 'b':0, 'l':0 
    "len": 0.9,    #the slider's length is this length minus the padding on both ends.  Default 1.
    "x": 0.1,   #number between or equal to -2 and 3. Sets the x position (in normalized coordinates) of the slider. Default 0.
    "y": 0,    #number between or equal to -2 and 3. Sets the y position (in normalized coordinates) of the slider. Default 0.
    "steps": [
        {"args": [
            [year],
            {"frame": {"duration": 300, "redraw": False},
             "mode": "immediate",
             "transition": {"duration": 300}}
        ],
        "label": year,
        "method": "animate"}
        for year in years
    ]
}
fig_dict["layout"]["sliders"] = [sliders_dict]

# make data
# 5 traces are appended into figure["data"]. One trace for each continent in year 1952. 
# Bunun yarari trace'lere isim vererek hangi rengin hangi kitaya ait oldugunu gosterebilmek.
year = 1952
for continent in continents:
    dataset_by_year = dataset[dataset["year"] == year]
    dataset_by_year_and_cont = dataset_by_year[
        dataset_by_year["continent"] == continent]

    #trace
    data_dict = {
        "x": list(dataset_by_year_and_cont["lifeExp"]),
        "y": list(dataset_by_year_and_cont["gdpPercap"]),
        "mode": "markers",
        "text": list(dataset_by_year_and_cont["country"]),
        "marker": {
            "sizemode": "area",  # one of ( "diameter" | "area" ) Default "diameter". Has an effect only if `marker.size` is set to a numerical array. 
            "sizeref": 200000,   # Sets the scale factor used to determine the rendered size of marker points. Use when `marker.size` is set to a numerical array.
            "size": list(dataset_by_year_and_cont["pop"])
        },
        "name": continent
    }
    fig_dict["data"].append(data_dict)

# make frames
for year in years:
    frame = {"data": [], "name": str(year)}
    for continent in continents:
        dataset_by_year = dataset[dataset["year"] == int(year)]
        dataset_by_year_and_cont = dataset_by_year[
            dataset_by_year["continent"] == continent]

        data_dict = {
            "x": list(dataset_by_year_and_cont["lifeExp"]),
            "y": list(dataset_by_year_and_cont["gdpPercap"]),
            "mode": "markers",
            "text": list(dataset_by_year_and_cont["country"]),
            "marker": {
                "sizemode": "area",
                "sizeref": 200000,
                "size": list(dataset_by_year_and_cont["pop"])
            },
            "name": continent
        }
        frame["data"].append(data_dict)

    fig_dict["frames"].append(frame)


fig = go.Figure(fig_dict)

fig.show()
```

```python
years=[i for i in data.year.unique()]
continents=[i for i in data.continent.unique()]


start_data=[
    go.Scatter(
        x=list(data[(data.year==1952) & (data.continent==continent)]["lifeExp"]),
        y=list(data[(data.year==1952) & (data.continent==continent)]["gdpPercap"]),
        text=list(data[(data.year==1952) & (data.continent==continent)]["country"]),
        mode='markers',
        marker=dict(
            size=list(data[(data.year==1952) & (data.continent==continent)]["pop"]),
            sizeref=200000,
            sizemode='area'
        ),
        name=continent
    )
    for continent in continents
]


layout=go.Layout(
    title='Life Expectancy vs GDP per Capita',
    xaxis=dict(range=[data["lifeExp"].min(), data["lifeExp"].max()*1.05], title='Life Expectancy'),
    yaxis=dict(type='log', title='GDP per Capita'),
    hovermode='closest',
    updatemenus=[
        dict(
            type='buttons',
            buttons=[
                dict(
                    args=[None, dict(frame=dict(duration=500, redraw=False), 
                                     fromcurrent=True,
                                     transition=dict(duration=300, easing='quadratic-in-out'))],
                    label='Play',
                    method='animate'
                ),
                dict(
                    args=[[None], dict(frame=dict(duration=0, redraw=False),
                                      mode='immediate',
                                      transition=dict(duration=300))],
                    label='Pause',
                    method='animate'
                )
            ],
            visible=True,
            showactive=False,
            direction='left',
            x=0.1,
            xanchor='right',
            y=0,
            yanchor='top',
            pad=dict(t=80, r=10)
        )
    ],
    sliders=[dict(       
        steps=[
            dict(
                args=[[year], dict(frame=dict(duration=300, redraw=False),
                              mode='immediate',
                              transition=dict(duration=300))],
                method='animate',
                label=str(year)
            ) 
            for year in years
        ],
        len=0.9,
        x=0.1,
        pad=dict(t=50, b=10),
        xanchor='left',
        y=0,
        yanchor='top',
        transition=dict(duration=300, easing='cubic-in-out'),
        currentvalue=dict(font=dict(size=20), prefix='Year:', visible=True, xanchor='right'),
    )]
)


frames=[
    go.Frame(
        name=str(year),
        data=[go.Scatter(
            x=list(data[(data.year==year) & (data.continent==continent)]["lifeExp"]),
            y=list(data[(data.year==year) & (data.continent==continent)]["gdpPercap"]),
            text=list(data[(data.year==year) & (data.continent==continent)]["country"]),
            mode='markers',
            marker=dict(
                size=list(data[(data.year==year) & (data.continent==continent)]["pop"]),
                sizeref=200000,
                sizemode='area'
            ),
            name=continent
        ) for continent in continents]
    )
    for year in years
]


fig=go.Figure(data=start_data, layout=layout, frames=frames)
fig.show()
```
